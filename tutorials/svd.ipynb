```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg
```

```python
img = cv2.imread("animal_lesser_panda_stand.png", cv2.IMREAD_GRAYSCALE)
U, S, V = np.linalg.svd(img)

fig = plt.figure(figsize=(16, 10))
ax = fig.add_subplot(2, 4, 1)
ax.imshow(img, cmap="gray")
ax.set_title(f"\nActual Image", fontsize=24)
ax.tick_params(labelbottom=False, labelleft=False, bottom=False, left=False)

for i, k in enumerate([1, 3, 5, 10, 25, 50, 364]):

    U_k = U[:, :k]
    S_k = np.matrix(linalg.diagsvd(S[:k], k, k))
    V_k = V[:k, :]

    reconstructed_img = U_k @ S_k @ V_k
    ax = fig.add_subplot(2, 4, i + 2)
    ax.imshow(reconstructed_img, cmap="gray")
    ax.set_title(f"\nk = {k}", fontsize=24)
    ax.tick_params(labelbottom=False, labelleft=False, bottom=False, left=False)

plt.tight_layout()
plt.show()
```
